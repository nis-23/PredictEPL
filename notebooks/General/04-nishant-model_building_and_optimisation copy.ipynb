```python
import pandas as pd 
import pickle

# Load up the datasets 

matches_v2 = pd.read_pickle('../data/interim/matches_interim_v2.pkl')
matches_v2_scaled = pd.read_pickle('../data/interim/matches_interim_v2_scaled.pkl')
matches_v3 = pd.read_pickle('../data/interim/matches_interim_v3.pkl')
matches_v3_scaled = pd.read_pickle('../data/interim/matches_interim_v3_scaled.pkl')

```

```python
# Training set 
# matches_v2


matches_v2['Year'] = pd.to_datetime(matches_v2.datetime).dt.year
matches = matches_v2


years = [2020, 2021]
matches2021 = matches.query('Year in  @years')

X_train = matches2021.drop(['target','datetime','team', 'season','opponent', 'Year'], axis = 1)
Y_train = matches2021['target']

# Dev Set & Test set 

matches2022 = matches.query('Year == 2022')

matches2022.drop(['datetime','team','season','opponent', 'Year'], axis = 1, inplace = True)

from sklearn.model_selection import train_test_split

X_test, X_dev, Y_test, Y_dev =  train_test_split(matches2022.drop('target', axis = 1), matches2022['target'],
                                                 test_size = 0.50, shuffle = True, random_state = 42)


# model_v2_scaled
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42, max_iter= 1000).fit(X_train, Y_train)

model_v2_preds = lr.predict(X_test)

from sklearn.ensemble import RandomForestClassifier 

rf = RandomForestClassifier(max_depth=10, random_state=42)
rf.fit(X_train, Y_train)

model_v2_preds_rf = rf.predict(X_test)

from sklearn.metrics import precision_score

precision_lr = precision_score(Y_test, model_v2_preds)
precision_rf =  precision_score(Y_test, model_v2_preds_rf)

precision_lr, precision_rf
```
