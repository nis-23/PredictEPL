```python
import pandas as pd 
import pickle

# Load up the datasets 

matches_v2 = pd.read_pickle('../data/interim/matches_interim_v2.pkl')
matches_v2_scaled = pd.read_pickle('../data/interim/matches_interim_v2_scaled.pkl')
matches_v3 = pd.read_pickle('../data/interim/matches_interim_v3.pkl')
matches_v3_scaled = pd.read_pickle('../data/interim/matches_interim_v3_scaled.pkl')
```

```python
# Training set 
# matches_v2



matches = matches_v2

matches2021 = matches.query('season == 2021')

X_train = matches2021.drop(['target','datetime','team', 'season','opponent'], axis = 1)
Y_train = matches2021['target']

# Dev Set & Test set 

matches2022 = matches.query('season == 2022')

matches2022.drop(['datetime','team','season','opponent'], axis = 1, inplace = True)

from sklearn.model_selection import train_test_split

X_test, X_dev, Y_test, Y_dev =  train_test_split(matches2022.drop('target', axis = 1), matches2022['target'],
                                                 test_size = 0.50, shuffle = True, random_state = 42)


# model_v2_scaled
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42, max_iter= 1000).fit(X_train, Y_train)

model_v2_preds = lr.predict(X_test)

from sklearn.metrics import precision_score

precision_score(Y_test, model_v2_preds)
```

```python
matches2022 = matches.query('season == 2022')
combined = pd.concat([matches2022,  pd.Series(model_v2_scaled_preds) ], axis=1)
new_columns = combined.columns.tolist()
new_columns[-1]  = 'Prediction'
combined.columns = new_columns
combined
```

```python
class MissingDict(dict):
    
    __missing__: lambda self, key: key
    
map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd",
        "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham",
        "Wolverhampton Wanderers": "Wolves"}


mapping = MissingDict(**map_values)

combined['new_team'] = combined['team'].map(mapping)
```
